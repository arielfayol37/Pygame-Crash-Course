We will move our Rect object that we called 'rect_object' to the right, then draw a rectangle at the new location of our Rect object.


When you execute pygame.display.update(), the new red rectangle will be drawn while the old red rectangle is still there.
Because the new image created by update() is actually a copy of the image from the previous update() call plus whatever you added to it.
For example if you don't add anything and call update(), nothing would have seem to change but what you would be seeing
is a different image but the exact copy of the previous image.

You may be thinking,"if I 'removed' the previous red rectangle and drew the new one where some pixel-distance away, it would appear as though the original rectangle has moved." Yeah, that's exactly what we do for animations. Create images and stack upon
others with seemingly smooth changes. Let's put that in code.


```python
screen.fill(screen_color)
#We repaint our screen to 'erase' the previous red rectangle or whatever there was on our screen
rect_object.x += 100 #we move our Rect object 100 pixels to the right
pygame.draw.rect(screen, rectangle_color, rect_object)#we draw a rectangle at the location of our 
#rect object
pygame.display.update()#we stack the new image on top of the old one
```

# Making the ball move

This is very easy if you understand what is going on so far. 
We are going to make our imaginary rectangle move, and draw our circle
each time inside it(or at its location to be precise).
Let's make it move towards the right non-stop

```python

x_velocity = 1.5
while True:
    circle_rect.x += x_velocity #increase the x-coordinate of the top left corner by 1(x_velocity)
                            
    screen.fill(screen_color) #repaint the screen (or wall)
    pygame.draw.circle(screen, circle_color, (circle_rect.x,circle_rect.y), radius) #draw the circle in the Rect
    pygame.display.update()#generate a new image with all what we have added and display
```

```python

import pygame
pygame.init()


screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width,screen_height))0


screen_color = (20,144,144)
screen.fill(screen_color)


###Creating a Rect object(region in space) where our circle will be.###
# The Rect object and the circle are indpendent please!
#The circle is not even an object at all. We just add the genie to draw a circle at location we indicate.
# 'drawing' actually is just changing the pixels of a defined region to whatever color we passed as argument.
#So there is nothing like a circle object that we can move. The only thing we move is our Rect object, and we keep
#drawing circles at its center
#We just use Rect object to facilitate animation
x, y = (55,90)
radius = 50
rect_width, rect_height = radius, radius
circle_color = (50,70,70)
circle_rect = pygame.Rect(x,y, rect_width, rect_height)
#x is the x-coordinate of the top-left corner of the Rect object, while y is the y-coordinate
#we set the dimesions of our Rect object to fit the circle as much as possible. 

#now drawing the circle:

pygame.draw.circle(screen, circle_color, (circle_rect.centerx,circle_rect.centery), radius)
#x in the first line of the cell is used to create the Rect object. It serves as the x-coordinate
#of the top-left corner of the Rect object

#centerx is an attribute of the Rect object, which is the x-coordinate of the center of the rect object
#we want the center of our circle to be at the center of the Rect object, that's why we pass the
#coordinates of the center of our Rect object, as arguments to the function used to draw the circle




#making the ball move, bounce, and change direction
x_velocity =1.5
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()#more about this later. This is just so you don't have to press Ctrl + C to stop the progam
                        #each time. You can now close it normally like any program
             
            
    if circle_rect.x <= 0 or (circle_rect.x + circle_rect.width) >= screen_width:
        x_velocity *= -1 #I am using some syntactic sugar. you could write x_velocity = x_velocity * -1
    '''In the condition, we check whether the left border(x-coordinate) of the circle_rect is touching or has crossed
    the left border of the screen, or if the right border of the circle_rect has crossed or is touching the right
    border of the screen. If yes, then we reverse the direction of the velocity'''
    circle_rect.x +=x_velocity #change the x position of the rect
    screen.fill(screen_color) #repaint the screen (or wall)
    pygame.draw.circle(screen, circle_color, (circle_rect.x,circle_rect.y), radius) #draw the circle at the location of the Rect
    pygame.display.update()#generate a new image with all what we have added and display
```

# **Converting to appropriate units for our Genie**
**This section is important because our genie does not understand the units given to us**

Concerning our physics problem, we were given velocity in metres per second. But our genie measures distances in pixel. So we first of all have to convert metres to pixel.
  
  ***Metres to Pixels***

    Conversion Factor to Pixel(CFP) = Height of our pygame window in pixel/height of the scenery in real life.

Since our building is 10m high, then we can approximate the total height of the scenery to be twice that.

    Height of scenery = 20m.

We will define the height of our pygame window in the code. Nowadays most laptops support HD format, meaning they have a minimum resolution(dimensions of the screen/monitor in pixels) of 1366 by 768 pixels. So we will set the height of our pygame window to be a more than that. 

    Height of pygame window = 800px.  

Therefore our CFP is:

    CFP = 800px/20m = 40px/m

***Note that a 40px/m CFP is good no matter the height of the building because
our eyes cannot see really far. If we make our CFP a function of the height of a building, then our ball will be very small for tall buildings(yes, cause the building is tall and you can't see very far) or the ball will appear very big for small building. For different problems like motion of planet in orbits, you will use a different CFP. Just do the maths and try out different values. 

(skipable)Also if the object is moving towards or away from the viewer, then you can keep resizing the moving object(the ball in this case) by applying some physics(lenses) to know the size the object should 'appear'***


***Formula to find position***

From classical physics(or common sense), you know
that at a constant velocity, 

    1) position = initial_position + velocity * time
                    
                            or

    2) curremt_position = previous_position + velocity * time_step


But in this context, there is acceleration, hence position as function of time is given by:

    position = initial_position + initial_velocity * time + 0.5 * acceleration * time^2

The velocity and acceleration are given to us in metres per second and metres per second per second respectively, so all we have to do is to convert to pixels per second and pixels per second per second by multiplying by our CFP.

There you go, now you can simulate free fall.

```python

import pygame #getting the tool box to build the game
import time #a tool we will use to track time. you don't need to install because
#it comes with python. It is an 'in-built' module.
pygame.init() #initializing(opening) the tool box

#Defining variables. Constants are going to be in upper case(programming conventions).
SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_COLOR = (1200,800, (255,255,255))
BALL_COLOR = (0,0,0)
building_height_in_m = 10
ball_radius_in_m = 0.2

#CFP = SCREEN_HEIGHT/(building_height_in_m * 2) #Conversion Factor to Pixel
CFP = 40
MASS = 3
BALL_RADIUS = ball_radius_in_m * CFP 
BUILDING_HEIGHT = building_height_in_m * CFP 
GRAVITY = 9.8 * CFP

initial_y_of_ball = SCREEN_HEIGHT - BUILDING_HEIGHT#We do this this way because of the orientation of our coordinate system
BUILDING_WIDTH = 100 #I chose this so that the ball starts some distance away from the border of the screen
initial_x_of_ball = BUILDING_WIDTH
initial_x_velocity, initial_y_velocity = tuple(map(lambda x: x * CFP, (5, -12)))#converting the 
#velocity from metres per second to pixels per seconds using the CFP.
#if you don't know this syntax, I am just saying, map this function to the two values (5,-12) an return the tuple.
#Google if you want to learn more.
#I could have just written: initial_x_velocity = 5 * CFP and initial_y_velocity = -12 * CFP


screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
#naming the window
pygame.display.set_caption('Free Fall')

#Creating the Rect object we will use for the ball
rect_width, rect_height = BALL_RADIUS , BALL_RADIUS
ball_rect = pygame.Rect(0,0,rect_width, rect_height)#we create the Rect object at a random position first. I chose (0,0) randomly
ball_rect.centerx, ball_rect.centery = initial_x_of_ball, initial_y_of_ball #I move the Rect object so that its center matches
#the initial position of the ball. Though I updated both the x and the y component, 
#I did not need to update both because when I update one, the other gets updated automatically

#Creating a rect for the building
building_rect = pygame.Rect(0, SCREEN_HEIGHT-BUILDING_HEIGHT, BUILDING_WIDTH, BUILDING_HEIGHT)
BUILDING_COLOR = (255,0,0)

initial_time = time.time()# returns time that has elapsed in seconds since 01/01/1970
time_elapsed = 0
#animation
while True:
    for event in pygame.event.get():#ignore this for now
        if event.type == pygame.QUIT:
            pygame.quit()
            
    time_elapsed = time.time() - initial_time
    ball_rect.centerx = initial_x_of_ball + initial_x_velocity * time_elapsed 
    ball_rect.centery = initial_y_of_ball + initial_y_velocity * time_elapsed + 0.5 * GRAVITY * time_elapsed ** 2
    screen.fill(SCREEN_COLOR) #repaint the screen (or wall)
    pygame.draw.circle(screen, BALL_COLOR, (ball_rect.centerx,ball_rect.centery), BALL_RADIUS) #draw the circle at the location of the Rect
    pygame.draw.rect(screen, BUILDING_COLOR, building_rect)
    pygame.display.update()#generate a new image with all what we have added and display
```

```python

import pygame #getting the tool box to build the game
import time
pygame.init() #initializing(opening) the tool box

#Defining variables. Constants are going to be in upper case(programming conventions).
SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_COLOR = (1200,800, (255,255,255))
BALL_COLOR = (0,0,0)
building_height_in_m = 10
ball_radius_in_m = 0.2

CFP = 40 #You can make the CFP a function of height so you can always see the entire building by uncommenting the next line
#CFP = SCREEN_HEIGHT/(building_height_in_m * 2) #Conversion Factor to Pixel
MASS = 3
BALL_RADIUS = ball_radius_in_m * CFP
BUILDING_HEIGHT = building_height_in_m * CFP 
GRAVITY = 9.8 * CFP

initial_y_of_ball = SCREEN_HEIGHT - BUILDING_HEIGHT#We do this this way because of the orientation of our coordinate system
BUILDING_WIDTH = 100 #I chose this so that the ball starts some distance away from the border of the screen
initial_x_of_ball = BUILDING_WIDTH
initial_x_velocity, initial_y_velocity = tuple(map(lambda x: x * CFP, (5, -12)))#converting the 
#velocity from metres per second to pixels per frames using the formula we derived earlier.
#if you don't know this syntax, I am just saying, map this function to the two values (5,-12) an return the tuple.
#Google if you want to learn more.
#I could have just written: initial_x_velocity = 5 * CFP/FRAME_RATE and initial_y_velocity = -12 * CFP/FRAME_RATE


screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
rect_of_screen = screen.get_rect()
#naming the window
pygame.display.set_caption('Free Fall')

#Creating the Rect object we will use for the ball
rect_width, rect_height = BALL_RADIUS, BALL_RADIUS
ball_rect = pygame.Rect(0,0,rect_width, rect_height)#we create the Rect object at a random position first. I chose (0,0) randomly
ball_rect.centerx, ball_rect.centery = initial_x_of_ball, initial_y_of_ball #I move the Rect object so that its center matches
#the initial position of the ball. Though I updated both the x and the y component, 
#I did not need to update both because when I update one, the other gets updated automatically

#Creating a rect for the building
building_rect = pygame.Rect(0, SCREEN_HEIGHT-BUILDING_HEIGHT, BUILDING_WIDTH, BUILDING_HEIGHT)
BUILDING_COLOR = (220,220,220)


animation_active = False
just_activated = False
play_writer = pygame.font.SysFont('Times New Roman', 48) #we use a specific writer for the play button because it will have a different
#font size from the tracking button
play_color = (0,0,0)
play_text = 'PLAY'
play_button_background_color = (220,220,220)
play_button_surface = play_writer.render(play_text, True, play_color, play_button_background_color)
#The last argument for background color is optional, or in python terms 'non_positional.'
rect_of_play_button_surface = play_button_surface.get_rect()
rect_of_play_button_surface.center = rect_of_screen.center #yeah, you don't need to specifiy x or y, it will take care of everything

track_writer = pygame.font.SysFont('Verdana', 15) #You can use any font. if the font(handwriting) you input is not a system font
#then the genie will use its defualt font.
track_color = (0,0,0)
track_text = '' #we will be determined below

RANGE = 'TBD' #To be determined



animation_just_started = True
flight_time = 0#used to count the amount of time the ball has been in the air

#animation
while True:
    
   
    for event in pygame.event.get():#ignore this for now
        if event.type == pygame.QUIT:
            pygame.quit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                animation_active = False   
                pygame.mouse.set_visible(True)
        elif event.type == pygame.MOUSEBUTTONDOWN:
            mouse_position = pygame.mouse.get_pos()
            if rect_of_play_button_surface.collidepoint(mouse_position) and animation_active == False:
                animation_active = True
                pygame.mouse.set_visible(False)
                previous_time = time.time()              
                 
                 


    if animation_active:
          
        
        flight_time += time.time() - previous_time
        previous_time = time.time()
        ball_rect.centerx = initial_x_of_ball + initial_x_velocity * flight_time
        ball_rect.centery = initial_y_of_ball + initial_y_velocity * flight_time + 0.5 * GRAVITY * flight_time** 2
        if ball_rect.centery >= SCREEN_HEIGHT and animation_just_started==False:
            animation_active = False
            
        animation_just_started= False
        
    RANGE = round((ball_rect.centerx-BUILDING_WIDTH)/CFP,2) #Converting back to metres and rounding to two decimal places.       
    x_position = round((ball_rect.centerx-BUILDING_WIDTH)/CFP, 2)
    y_position = round((SCREEN_HEIGHT-ball_rect.centery-BUILDING_HEIGHT)/CFP, 3)
    
    track_text = f'X: {x_position} m            Y: {y_position} m       RANGE: {RANGE} m      time:{round(flight_time, 3)}s'
    track_surface = track_writer.render(track_text,True, track_color)
    screen.fill(SCREEN_COLOR) #repaint the screen (or wall)
    if animation_active == False:
        screen.blit(play_button_surface, rect_of_play_button_surface)
    screen.blit(track_surface, (0,0))#the blit method can also take the coordinate of the top left corner of where you want to display the surface
    pygame.draw.circle(screen, BALL_COLOR, (ball_rect.centerx,ball_rect.centery), BALL_RADIUS) #draw the circle at the location of the Rect
    pygame.draw.rect(screen, BUILDING_COLOR, building_rect)
    pygame.display.update()#generate a new image with all what we have added and display
    

```
